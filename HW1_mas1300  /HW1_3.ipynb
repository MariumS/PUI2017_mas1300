```python
import os
import sys

# you can use the numpy library for random number generation: 
# http://docs.scipy.org/doc/numpy/reference/routines.random.html

import numpy as np


np.random.seed(11)

ReprRand= np.random.randn(2,100)
ReprRand.shape
```

```python
ReprRandAll = np.zeros((50, 2, 100))
stdd = np.random.randint(2,100)
meann = np.random.randint(2,100)
for i in range(50):
    ReprRandAll[i]= stdd * np.random.randn(2, 100) + meann
    #print ReprRandAll
    #code here...
```

```python
ReprRandAll.shape
```

```python
import pylab as pl
%pylab inline

pl.xlabel('x')
pl.ylabel('y')
pl.title('pseudo random')
pl.grid(True)

for i in range(50):
    for t in range(100):
        pl.plot(ReprRandAll[i][0][t], ReprRandAll[i][1][t], 'o')
        
pl.show()
```
