```python
'''new dataset only with columns required'''
df1= df[['NUMBER OF PEDESTRIANS INJURED','NUMBER OF MOTORIST INJURED']].copy()
df1.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

'''lighter dots indicate less dots layered over one another, meaning less overlap of number of pedestrians injured and
number of motorists injured at that point '''

%matplotlib inline

y = df1['NUMBER OF PEDESTRIANS INJURED']
x = df1['NUMBER OF MOTORIST INJURED']
colors = ('pink')
plt.ylabel('number of pedestrians injured')
plt.xlabel('number of motorists injured')
plt.title('Relative Number of Injuries in a Motor Vehicle Collision')
#plt.legend("one dot= one injury")


classes = ['lighter=less dots overlapped']
class_colours = ['pink']
recs = []
for i in range(0,len(class_colours)):
    recs.append(mpatches.Circle((0,1),fc=class_colours[i]))
plt.legend(recs,classes,loc=1)

plt.scatter(x, y, c=colors, alpha=0.19)
plt.show()
```
